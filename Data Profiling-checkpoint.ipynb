```python
#Suppress warning messages
import warnings
warnings.filterwarnings("ignore")
import os
import gc
import itable
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# plt.style.use('fa')
import pylab
import time
# from Tkinter import *
from prettytable import PrettyTable
# from tkFileDialog import askopenfilename
from IPython.core.display import display, HTML
%matplotlib inline
pylab.rcParams['figure.figsize'] = (9, 7)

pd.set_option('display.width', 1000)

#Set font size for matplotlib
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 20}

plt.rc('font', **font)
plt.rc('xtick', labelsize = 10)
plt.rc('ytick', labelsize = 10)
```

# File Properties

```python
filename = 'Mortality_Drv.csv'
dat = pd.read_csv(filename, sep = ',')

display(HTML('<h4>Filename: '+os.path.split(filename)[1]+'</h4>'))
display(HTML('<h4>Observation: '+str(dat.shape[0])+'</h4>'))
display(HTML('<h4>Column: '+str(dat.shape[1])+'</h4>'))
display(HTML('<h4>Last modified: '+time.ctime(os.path.getmtime(filename))+'</h4>'))
display(HTML('<h4>Created: '+time.ctime(os.path.getctime(filename))+'</h4>'))

col_names = dat.columns
col_types = list(dat.dtypes)
```

# Warnings

```python
sum_alert = pd.DataFrame(columns = ["Name", "Alert"], index = range(0,1))
flag = 0

for cn in col_names:
    alert = ""
    num_MissData = sum(dat[cn].isnull())
    if (num_MissData) / (float(dat.shape[0])) > 0.2:
        alert += "MISS "
    
    if dat[cn].dtype == int or dat[cn].dtype == float:
        outliers = dat[cn][dat[cn] > np.std(dat[cn]) * 5]
        if len(outliers) > 0:
            alert += "OUTLIER"

    if bool(alert):
        if flag == 0:
            sum_alert["Name"] = cn
            sum_alert["Alert"] = alert
            flag = flag + 1
        else:
            sa = pd.DataFrame(index = range(0,1))
            sa["Name"] = cn
            sa["Alert"] = alert
            sum_alert = sum_alert.append(sa)

# Reset index
sum_alert.index = [x for x in range(sum_alert.shape[0])]
if sum_alert.shape[0] != 0:
    display(HTML("<table style='width:100%'><tr><td style='color:red'><font size = 3> \
                 <b>MISS: More than 20% missing values.<br>OUTLIER: More than 5 standard deviations \
                 from the mean.</b></font></td></table>"))
else:
    display(HTML("<table style='width:100%'><tr><td style='color:red'><font size = 3><b>There are \
                 neither missing values nor outliers.</b></font></td></table>"))
itable.PrettyTable(sum_alert, tstyle=itable.TableStyle(theme="theme1"), center=True)
```

# Columns' Information 

```python
for i, dt in enumerate(col_types):
    colName = col_names[i]
    if dt == 'object':
        display(HTML('<table style="width:100%"><tr><td style="color:black"><h2>Column '+
                     str(i)+": "+colName+'</h2></td></table>')) 
        display(HTML('<h2>Type: Object</h2>'))
        display(HTML('<h3>1. Top 50 Most Frequent Values.</h3>'))
        freq = dat[colName].value_counts()
        fifty_freq = freq[:50]
        df = pd.DataFrame(fifty_freq.values, columns=["Freq"])
        df["Attribute Name"] = fifty_freq.index
        #Swap the columns order
        df2 = df[['Attribute Name', 'Freq']]    
        y = sum(freq)
        df["Percent of Total"] = [str(round((x/float(y))*100, 2))+"%" for x in fifty_freq]
        num_uniq = len(freq)
        df.index = [x for x in range(1,df.shape[0]+1)]
        print(df)
        display(HTML('<table><tr><td style="color:black"><h4>Unique Values: '+
                     str(num_uniq)+'</h4></td></tr></table>'))
        
        display(HTML('<h3>2. Value Distribution.</h3>'))
        f = plt.figure()
        ax = f.add_subplot(111)
        ax.yaxis.tick_right()
        temp = dat[colName].value_counts(sort = False)
        temp.plot(kind = "bar", alpha = 0.6)
        plt.title(colName+" Distribution", )
        plt.show()
        print("\n")
        
        display(HTML('<h3>3. Missing Values.</h3>'))
        if dat[colName].isnull().any():
            numMissingValues = sum(dat[colName].isnull())
            display(HTML('<h4>There are '+str(numMissingValues)+' missing values.</h4>'))
            ratioMissingOfTotal = round(numMissingValues / float(dat.shape[0]), 2)
            display(HTML(str(ratioMissingOfTotal * 100)+
                         '<h4>% of total variables are considered missing variables.</h4><br><br>'))
        else:
            display(HTML('<h4>There are no missing values!</h4><br><br>'))
    elif dt == 'int' or dt == 'float':
        display(HTML('<table style="width:50%"><tr><td style="color:black"><h2>Column '+
             str(i)+": "+colName+'</h2></td></table>')) 
        if dt == 'int':
            display(HTML('<h2>Type: Integer</h2>')) 
        elif dt == 'float':
            display(HTML('<h2>Type: Float</h2>'))
        
        display(HTML('<h3>1. Basic Data Statistics.</h3>'))
        display(HTML('<table><tr><td style="color:black">Count</td><td>'\
                     +str(len(dat[colName])-sum(dat[colName].isnull()))+'</td></tr>\
                     <tr><td>Mean</td><td>'+str(round(np.mean(dat[colName]),2))+'\
                     </td></tr><tr><td>Standard Deviation</td><td>'+str(round(np.std(dat[colName]),2))+'</td>\
                     </tr><tr><td>Min</td><td>'+str(round(np.min(dat[colName]),2))+'</td></tr><tr><td>Max</td>\
                     <td>'+str(round(np.max(dat[colName]),2))+'</td></tr></table>'))
        display(HTML('<h3>2. Value Distribution.</h3>'))
        if sum(dat[colName].isnull()) < dat.shape[0] / 2:
            f = plt.figure()
            ax = f.add_subplot(111)
            ax.yaxis.tick_right()  
            ax.xaxis.set_ticklabels([])
            plt.hist(dat[colName].dropna(), bins = 40, alpha = 0.6)
            plt.title(colName+" Distribution")    
            plt.show()
        else:
            f = plt.figure()
            ax = f.add_subplot(111)
            count_pct = dat[colName].dropna().value_counts(sort = False) 
            count_pct.plot(kind = "bar", alpha = 0.6)
            plt.title(colName+" Distribution")
            ax.yaxis.tick_right()
            ax.xaxis.set_ticklabels([])
            plt.show() 
        # Adjust boxplot sizes
        f2 = plt.figure(figsize = (8.9,3))
        ax2 = f2.add_subplot(111)
        ax2.yaxis.tick_right()            
        plt.show(dat.boxplot(column = [colName], vert = False)) 
        
        # Checking missing values
        display(HTML('<h3>3. Missing Values.</h3>'))
        if dat[colName].isnull().any():
            numMissingValues = sum(dat[colName].isnull())
            display(HTML('<h4>There are '+str(numMissingValues)+' missing values.</h4>'))
            ratioMissingOfTotal = round(numMissingValues / float(dat.shape[0]), 2)
            display(HTML('<h4>'+str(ratioMissingOfTotal * 100)+'% of total variables are \
                         considered missing variables.</h4><br><br>'))
        else:
            display(HTML('<h4>There are no missing values!</h4><br><br>'))
            
        # Detect Outliers (if above or below 4 standard deviation)
        outliers = dat[colName][dat[colName] > np.std(dat[colName]) * 4]
        display(HTML('<h3>4. Outliers.</h3>'))
        if len(outliers) == 0:
            display(HTML('<h4>There are no outliers!</h4><br><br>'))
        else:
            outlierRatio = round(len(outliers) / float(dat.shape[0]), 2)
            display(HTML('<h4>'+str(len(outliers))+' outliers are detected.</h4>'))
            display(HTML('<h4>'+str(outlierRatio * 100)+'% of variables are considered \
                         outliers.</h4><br><br>'))
```

# Missing Data and Patterns

<table style="width:100%">
<tr><td style="color:black"><center><font size = 5><b>Basic Columns Summary with Missing Data</b></font></center></td>
</table>

```python
#Missing value basic info
temp = pd.DataFrame(columns = ["Name", "Num_Missing", "Percentage_Of_Missing"],
                   index = range(0,1))

flag = 0
for cn in col_names:
    num_MissData = sum(dat[cn].isnull())
    if num_MissData != 0:
        
        if flag == 0:
            temp["Name"] = cn
            temp["Num_Missing"] = num_MissData
            temp["Percentage_Of_Missing"] = str(round((num_MissData / float(dat.shape[0])), 4) * 100)+"%"
            flag = flag + 1
        else:
            tp = pd.DataFrame(index = range(0,1))
            tp["Name"] = cn
            tp["Num_Missing"] = num_MissData
            tp["Percentage_Of_Missing"] = str(round((num_MissData / float(dat.shape[0])), 4) * 100)+"%"
            temp = temp.append(tp)

temp = temp.sort(["Num_Missing"], ascending=False)
temp.index = [x+1 for x in range(temp.shape[0])]
manyWarning = temp[temp.Num_Missing > dat.shape[0]/2]
tb = itable.PrettyTable(temp, tstyle=itable.TableStyle(theme="theme1"), center=True)
cs = itable.CellStyle()
cs.set("background-color", "red")
cs.set("color", "white")
tb.set_cell_style(style=cs, rows=manyWarning.index-1, cols=[0,1,2])
tb
```

<table style="width:100%">
<tr><td style="color:black"><center><font size = 5><b>Missing Data Patterns</b></font></center></td>
</table>

```python
#Missing value pattern and its index
temp2 = pd.DataFrame(columns = ["Pattern", "Pattern_Type", "Percent of Rows"], index = range(0,1))

flag = 0
patt_idx = 1
#Create a dataframe containing only missing values
null_data = dat[dat.isnull().any(axis=1)]
nan_patt = null_data.isnull()
for npv in nan_patt.values:
    #Create a missing value pattern
    patt_str = ""
    for p in npv:
        if p:
            patt_str = patt_str + "|"
        else:
            patt_str = patt_str + "."

    if flag == 0:
        temp2["Pattern"] = patt_str
        temp2["Pattern_Type"] = patt_idx
        patt_idx += 1
        flag += 1
    else:
        pp = pd.DataFrame(index = range(0,1))
        pp["Pattern"] = patt_str
        if patt_str not in temp2.Pattern.values:
            pp["Pattern_Type"] = patt_idx
            patt_idx += 1
        else:
            ix = pp["Pattern"][0] == temp2.Pattern.values
            dup_patt_idx = np.where(ix == True)[0][0]
            pp["Pattern_Type"] = temp2["Pattern_Type"].iloc[dup_patt_idx]
        temp2 = temp2.append(pp)

# Pattern and its type number
xx = pd.DataFrame({'Count' : temp2.groupby( ["Pattern", "Pattern_Type"] ).size()}).reset_index()
xx["Percent of Rows"] = list(map(lambda x: str(round(x,2))+"%", (xx["Count"] / dat.shape[0])*100))
#Sort by Count
xx = xx.sort(["Count"], ascending=False)

# Rearrange columns
xx = xx[["Pattern_Type", "Pattern", "Count", "Percent of Rows"]]
xx.Pattern_Type = [z for z in range(1, xx.shape[0]+1)]

#Figure out the number of rows of full data
full_row = round((dat.shape[0] - sum(xx['Count'])) / dat.shape[0], 2)
display(HTML('<h3>Percentage of full data: '+str(full_row)+'%</h3>'))
itable.PrettyTable(xx.iloc[:20], tstyle=itable.TableStyle(theme="theme1"), center=True)
```

```python
if xx.shape[0] >= 20:
    mfp_df = pd.DataFrame(columns = ["Pattern_Type", "Count", "Percent of Rows", "Variables Missing"],
                     index = range(0,20))
else:
    mfp_df = pd.DataFrame(columns = ["Pattern_Type", "Count", "Percent of Rows", "Variables Missing"],
                 index = range(0,xx.shape[0]))
    
# If there are less than 20 missing data pattern
if mfp_df.shape[0] < 20:
    top_n_pattern_type_cnt = xx["Count"][:xx.shape[0]]
    pt = xx["Pattern_Type"].values[:xx.shape[0]]
# If there more than 20 missing data pattern, show the top 20 missing data pattern
else:
    top_twenty_pattern_type_cnt = xx["Count"][:20]
    pt = xx["Pattern_Type"].values[:20]

perc_row = ((top_twenty_pattern_type_cnt / dat.shape[0]).values) * 100
perc_row = map(lambda x: round(x,2), perc_row)

mfp_df["Pattern_Type"] = pt
mfp_df["Count"] = top_twenty_pattern_type_cnt.values
mfp_df["Percent of Rows"] = list(map(lambda x: str(x)+"%", perc_row))

for j, p in enumerate(xx.Pattern.values):
    vars_miss = ""
    for i, patt in enumerate(p):
        if patt == "|":
            vars_miss += col_names[i]+" // "
    mfp_df["Variables Missing"].loc[j] = vars_miss

# Sort by the number of rows
mfp_df = mfp_df.sort(["Count"], ascending=False)

mfp_df.index = [x+1 for x in range(mfp_df.shape[0])]
tb = itable.PrettyTable(mfp_df, tstyle=itable.TableStyle(theme="theme1"), center = True)
tb
```
